```python
# Ödenen ücreti 40'dan büyük olan satırları getirin
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
df[df.total_bill > 40]
```

```python
# Soru : Cinsiyeti erkek olan satırlar
```

```python
df[df.sex == 'Male']
```

```python
# Soru : Cinsiyeti kadın ve Bahşişi 5'ten büyük olan satırları göster
```

```python
df[(df.sex == 'Female') & (df.tip > 5)] # SORGULARIN ARASINA & SEMBOLU VE SORGULAR PARANTEZL OLSUN
```

```python
# Soru : Günü Pazar(Sun) olan VE ödenen miktar 30'dan büyük olan satırları getirin.
```

```python
df[(df.day == 'Sun') & (df.total_bill>30)]
```

```python
# Günü Cumartesi veya Pazar olanlar
```

```python
df[(df.day == 'Sat') | (df.day=='Sun')]
```

```python
# Günü cumartesi veya pazar veya perşembe olanlar
```

```python
df[(df.day == 'Sat') | (df.day=='Sun') | (df.day == 'Thu')]
```

```python
df.day.isin(['Sat','Sun','Thu'])
```

```python
df[df.day.isin(['Sat','Sun','Thu'])]
```

```python
# Günü Cumartesi olmayanlar
```

```python
df[df["day"] != 'Sun']
```

```python
df[~(df["day"] != 'Sun')]
```
