# .apply() Methodu

```python
# .apply() bir series ya da dataframe sütunlarındaki değerlerin hepsine bir anda bir fonksiyonu uygulama yarar.
# Arka planında for döngüsü çalışır dolayısıyla çok hızlı değildir performans açısından vektörize edilmiş işlemleri uygulamak daha faydalıdır.
# Vektörize edilmiş işlemler oldukça hızlı çalışır ve for döngüsü kullanımını minimize eder.
# Ancak diğer taraftan .apply() methodunu kullanmak oldukça okunaklıdır. Başlangıçta kolaydır ve her durumu uygulanabilir.
# Ancak vektörize işlemler bazı durumlarda yapılamayabilir
```

```python
# Amaç : CC Number sütunundaki değerlerin son 4 rakamını almak ve CCNumber sütununu bu şekilde güncellemek. (CC Number -> int64)
```

```python
# Amaç : 10 dolardan az ödediyse $, 10 dolar ile 30 dolar arasında ödediyse $$, 30 dolardan fazla ödediyse $$$ işareti bulunan fiyat kategorisi adında bir sütun oluşturun
```

# 1. Yol

```python
def fiyat_kategori(x):
  if x<10:
    return '$'
  elif x>=10 and x<=30:
    return '$$'
  else:
    return '$$$'    
```

```python
df["Fiyat Kategori"] = df['total_bill'].apply(fiyat_kategori)
```

```python
df.sample(5)
```

# 2. Yol

```python
df.head()
```

```python
df["Fiyat Kategori"] = df["total_bill"].apply(lambda x : '$' if x < 10 else ('$$' if 10 <= x <=30 else '$$$'))
```

```python
df.sample(5)
```

# 3.Yol (Vektörize)

```python
df.head()
```

```python
def fiyat_kategori_vectorized(bill):
  
  categories = np.full_like(bill , '$' ,dtype= 'U3')

  categories[(bill >= 10) & (bill <= 30)] = '$$'

  categories[bill > 30] = '$$$'

  return categories
```

```python
df["Fiyat Kategori"] = fiyat_kategori_vectorized(df["total_bill"].values)
```

```python
df.sample(5)
```
