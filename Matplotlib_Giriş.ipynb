```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12, 8))
```

```python
plt.show()
```

# plt.plot Fonksiyonu

```python
# f(x) = 2x
x = np.arange(0,7)
y = np.arange(0,13,2)

plt.figure(figsize = (5, 5))

plt.plot(x, y)

plt.show()
```

### Soru : f(x) = -3x fonksiyonunu çizdiriniz.

```python
x = np.arange(0,7)
y = -3 * x

plt.plot(x, y)

plt.show()
```

```python
x = np.arange(0,7)
y = -3 * x

plt.figure(figsize = (5, 5))

plt.plot(x, y, '--', color = "green")

plt.show()
```

```python
x = np.arange(0,7)
y = 4 * x - 2

plt.figure(figsize = (5, 5))

plt.plot(x, y, 'r+', color = "red")

plt.show()
```

### Soru : f(x) = 2x+3 fonksiyonunu daire olarak çizdiriniz.

```python
x = np.arange(7)
y = 2 * x - 3

plt.plot(x, y, 'o')
```

# Lineer Olmayan Fonksiyonlar

```python
x = np.arange(100) / 10
y = np.sin(x)

plt.plot(x, y)
```

# Etiket Koyma ve Grafik Üzerinde Görüntüleme

```python
x = np.arange(6)
y1 = x * 2
y2 = x ** 2 

plt.plot(x , y1, label = '2x')
plt.plot(x , y2, label = 'x^2')

plt.legend()
```

# plt.scatter() - Nokta Grafiği

```python
x = np.arange(6)
y = x * 2

plt.scatter(x, y, color='b')
plt.plot(x, y, color = 'r')
```

```python
# Fonksiyonda karşılığı olan değerleri nokta içine alır.
```

# Soru : sin(x) plot, cos(x) scatter olacak şekilde grafik çizdiriniz.

```python
x = np.arange(100) / 10
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sinx' , color = 'r')
plt.scatter(x, y2, label = 'cosx', color = 'y')

plt.legend()
```

# plt.plot ve plt.scatter Parametreleri

--plt.plot


* 'color' = çizgi rengi
* 'linestyle' = çizgi stili
* 'marker' = grafikte karşılığı gelen yere noktanın stili
* 'markersize' = nokta boyutu
* 'alpha' = saydamlık
* 'label' = etiket plt.legend() ile de görüntülenir



```python
x = np.arange(100) / 10
y1 = np.sin(x)

plt.plot(x, y1, label = 'sinx' , color = 'g', linestyle = '--')

plt.legend()
```

```python
plt.figure(figsize = (6,4))

x = np.arange(100)

plt.plot(x, np.sin(x), marker='o', markersize = 2, alpha = 0.5)
```

--plt.scatter


* s = (size) boyut
* edgecolor = noktayı saran renk
* linewidth = noktayı saran kenarlığın ebatı

```python
x = np.arange(100) / 10
y = np.cos(x)

plt.scatter(x, y, label = 'cosx', color = 'black',s=100, edgecolor='red', linewidth = 2)

plt.legend()
```
