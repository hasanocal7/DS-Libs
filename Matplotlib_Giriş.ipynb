# plt.plot Fonksiyonu

```python
# f(x) = 2x
x = np.arange(0,7)
y = np.arange(0,13,2)

plt.figure(figsize = (5, 5))

plt.plot(x, y)

plt.show()
```

### Soru : f(x) = -3x fonksiyonunu Ã§izdiriniz.

```python
x = np.arange(0,7)
y = -3 * x

plt.plot(x, y)

plt.show()
```

```python
x = np.arange(0,7)
y = -3 * x

plt.figure(figsize = (5, 5))

plt.plot(x, y, '--', color = "green")

plt.show()
```

```python
x = np.arange(0,7)
y = 4 * x - 2

plt.figure(figsize = (5, 5))

plt.plot(x, y, 'r+', color = "red")

plt.show()
```

# Etiket Koyma ve Grafik Ãzerinde GÃ¶rÃ¼ntÃ¼leme

```python
x = np.arange(6)
y1 = x * 2
y2 = x ** 2 

plt.plot(x , y1, label = '2x')
plt.plot(x , y2, label = 'x^2')

plt.legend()
```

# plt.scatter() - Nokta GrafiÄi

# Soru : sin(x) plot, cos(x) scatter olacak Åekilde grafik Ã§izdiriniz.

```python
x = np.arange(100) / 10
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sinx' , color = 'r')
plt.scatter(x, y2, label = 'cosx', color = 'y')

plt.legend()
```

# plt.plot ve plt.scatter Parametreleri

--plt.plot


* 'color' = Ã§izgi rengi
* 'linestyle' = Ã§izgi stili
* 'marker' = grafikte karÅÄ±lÄ±ÄÄ± gelen yere noktanÄ±n stili
* 'markersize' = nokta boyutu
* 'alpha' = saydamlÄ±k
* 'label' = etiket plt.legend() ile de gÃ¶rÃ¼ntÃ¼lenir



```python
x = np.arange(100) / 10
y1 = np.sin(x)

plt.plot(x, y1, label = 'sinx' , color = 'g', linestyle = '--')

plt.legend()
```

```python
plt.figure(figsize = (6,4))

x = np.arange(100)

plt.plot(x, np.sin(x), marker='o', markersize = 2, alpha = 0.5)
```

--plt.scatter


* s = (size) boyut
* edgecolor = noktayÄ± saran renk
* linewidth = noktayÄ± saran kenarlÄ±ÄÄ±n ebatÄ±

```python
x = np.arange(100) / 10
y = np.cos(x)

plt.scatter(x, y, label = 'cosx', color = 'black',s=100, edgecolor='red', linewidth = 2)

plt.legend()
```
