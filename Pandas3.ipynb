# Missing Value ile Yapılabilecek 3 Şey Var

```python
#1 Dokunmadan bırakılabilir.
#2 Drop etmek (Satır ve Sütun bazlı)
#3 Mantıklı bir şekilde doldurulabilir.
```

# Drop Etmek

```python
df.dropna()
```

```python
df.dropna(how = 'all') # Hepsi NaN olması şartıyla siler
```

```python
df.dropna(thresh = 1) # -> 1 tane bile NaN olmayan değer varsa tut
```

```python
df.dropna(axis = 1)
```

# Fill the Data (NaN Değeri Doldur)

```python
df
```

```python
df.dropna(how = 'all' , inplace = True)
```

```python
df.post_movie_score.fillna(9.5 , inplace = True)
```

```python
df
```

```python
df.pre_movie_score.fillna(df.pre_movie_score.mean())
```
